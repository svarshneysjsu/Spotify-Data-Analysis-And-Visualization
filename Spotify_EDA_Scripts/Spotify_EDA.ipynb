```python
# Converting song duration from millisec to mins 
songs['duration_mins'] = songs['duration_ms'] / (1000 * 60)
```

```python
# Group Years into Decades
songs['decade'] = (songs['year'] // 10) * 10
```

```python
# 3. Categorize `popularity` into High, Medium, Low
songs['popularity_category'] = pd.cut(songs['popularity'], bins=[0, 49, 79, 100], 
                                   labels=['Low', 'Medium', 'High'], include_lowest=True)
```
